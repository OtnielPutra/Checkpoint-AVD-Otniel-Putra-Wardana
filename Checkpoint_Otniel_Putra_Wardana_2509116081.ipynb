## Check Outliers

pada bagian ini, kita akan melihat apakah ada nilai anomali (outliers) pada kolom data. hal ini dilakukan agar kita mendapatkan data yang akurat sehingga kita dapat memperkuat analisis kita nanti. akan tetapi, tidak semua kolom perlu kita tangani. kita tetap harus melihat tujuan dari kolom tersebut untuk apa didalam dataset

terlihat bahwa hanya kolom `transaction_amount` yang memiliki outliers didalam data sekitar 4%

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## Mengecek Outliers

pada bagian ini, saya akan mengecek apakah ada outliers dalam data saya

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

#Dataframe dari list hasil

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

#Tampilkan dataframe

display(results_df)
```

terlihat bahwa di `transaction_amount` terdapat outliers sebesar 4,4% yang menandakan saya akan memperbaikinya dengan capping flooring

```python
columns_to_impute = ["transaction_amount"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

untuk membersihkannya, saya menggunakan kode diatas dan menulis pada kolom mana yang mau dibersihkan dari outliers

# Data Construct

```python
def kategori_transaksi(transaction_amount):
    if transaction_amount < 200:
        return 'rendah'
    elif transaction_amount <= 500:
        return 'Menengah'
    else:
        return 'tinggi'

df['kategoti_transaksi'] = df['transaction_amount'].apply(kategori_transaksi)
```

kali ini, saya akan menggunakan bar char yang dimana kegunaannya yaitu membandingkan penjualan tiap barang yang ada

```python
Top5Penjualan = df['item_name'].value_counts().head(5).sort_values(ascending=False) # head() jumlah yang mau ditampilkan & Urutan

plt.figure(figsize=(10, 6)) # Ukuran
plt.bar(Top5Penjualan.index, Top5Penjualan.values, color='skyblue') # color warna
plt.xlabel('Nama makanan/minuman') # keterangan X label
plt.ylabel('Banyaknya penjualan') # keterangan y label
plt.title('Top Penjualan Terbanyak') # Judul
plt.xticks(rotation=45, ha='right') # rotasi teks
plt.tight_layout()
plt.show()
```

**Insight** : Berdasarkan grafik, terlihat kalau Cold Coffee mendominasi dalam penjualan produk, dan diikuti oleh produk ;ainnya seperti sugarcane juice, panipuri, frankie, dan aalopuri. hal ini menunjukan bahwa Cold Coffee lebih banyak diminati daripada produk lainnya

**Action** : Tindakan yang dapat dilakukan adalah memanfaatkan menu yang paling tinggi seperti Cold Coffee sebagai produk unggulan, misalnya dengan melakukan promosi atau paket bundling dengan menu lain agar dapat meningkatkan penjualan produk yang nilainya lebih rendah seperti Frankie dan Aalopuri. Selain itu, evaluasi juga bisa dilakukan pada menu dengan nilai rendah untuk melihat apakah perlu perbaikan rasa, harga, atau strategi promosi agar lebih menarik bagi pelanggan.

## Line Chart

kali ini saya akan menggunakan line chart yang berguna untuk memberikan visual terhadap perkembangan penjualan setiap bulannya

```python
# Pastikan kolom bertipe datetime

# Group by month
penjualanSetiapBulan = df.groupby(df['date'].dt.to_period('M')).size()

plt.figure(figsize=(20, 6)) # Ukuran figure diperbesar untuk kejelasan
plt.plot(penjualanSetiapBulan.index.astype(str), penjualanSetiapBulan.values, marker='o', color='blue') # Mengubah warna agar berbeda
plt.title('Penjualan Seluruh Barang Berdasarkan Bulan', fontsize=16) # Judul
plt.xlabel('Tanggal (Bulan)', fontsize=12) # X Keterangan
plt.ylabel('Penjualan', fontsize=12) # Y Keterangan
plt.grid(True)
plt.gcf().autofmt_xdate() # Otomatis mengatur format dan rotasi di sumbu X
plt.tight_layout()
plt.show()
```

**Insight** : Terlihat dari data diatas, bahwa pada bulan 4 tahun 2022 adalah  penjualan terletak paling bawah daripada semmua data. di bulan selanjutnya meroket menjadi penjualan terbanyak diantara semua bulan yang ada. dan penjualan terus naik turun hingga bulan 11 terjadinya penjualan yang termasuk lumayan stabil dari sebelumnya. hal ini menunjukkan bahwa penjualan barang sering mengalami kenaikan dan penurunan yang drastis.

**Action**: Tindakan yang dapat dilakukan adalah dengan menganalisis lebih lanjut faktor yang menyebabkan peningkatan penjualan pada bulan-bulan tertentu, seperti promosi, musim, atau perilaku pelanggan. Dari situ, strategi yang sama bisa diterapkan kembali pada bulan dengan penjualan rendah, misalnya dengan memberikan promo khusus, diskon, atau paket menu untuk menarik lebih banyak pembeli agar penjualan lebih stabil sepanjang tahun.

## Histogram

```python
plt.figure(figsize=(8, 5))
plt.hist(df['quantity'], bins=20, color='blue', edgecolor='black')
plt.title('Distribusi Jumlah Item yang Dibeli per Transaksi')
plt.xlabel('Jumlah Item')
plt.ylabel('Frekuensi Transaksi')
plt.show()
```

**Insight** : pada kolom quantity digunakan untuk melihat distribusi jumlah item yang dibeli dalam setiap transaksi. Dari grafik ini dapat diketahui apakah sebagian besar pelanggan hanya membeli sedikit item (misalnya 1–2 produk) atau ada juga transaksi yang membeli dalam jumlah banyak. Hal ini membantu memahami pola pembelian pelanggan secara umum.

**Action** : Jika sebagian besar transaksi hanya membeli sedikit item, maka penjual bisa membuat promo bundling atau paket hemat agar pelanggan tertarik membeli lebih banyak produk sekaligus. Dengan strategi ini, jumlah item per transaksi dapat meningkat sehingga total penjualan juga berpotensi bertambah.

## Boxplot

```python
VARIABEL1 = df['item_name'].value_counts().head(5).index
VARIABEL2 = df[df['item_name'].isin(VARIABEL1)]

plt.figure(figsize=(12, 7)) # Ukuran figure
VARIABEL2.boxplot(column='quantity', by='item_name', ax=plt.gca(), grid=False, patch_artist=True, medianprops=dict(color='red'))
plt.title('Distribusi Jumlah Pembelian pada 5 Menu Terpopuler', fontsize=16) # Judul
plt.xlabel('Nama Menu', fontsize=12) # X Keterangan
plt.ylabel('Quantitas', fontsize=12) # Y Keterangan
plt.xticks(rotation=45, ha='right')
plt.suptitle('') # Menghilangkan judul default dari boxplot.by
plt.tight_layout()
plt.show()
```

**Insight:**
Boxplot ini menunjukkan distribusi jumlah pembelian (quantity) pada 5 menu yang paling sering muncul dalam data. Dari grafik tersebut dapat dilihat bagaimana sebaran jumlah pembelian pada tiap menu, termasuk nilai median, rentang pembelian, serta kemungkinan adanya transaksi yang membeli dalam jumlah jauh lebih banyak dari biasanya (outlier). Hal ini membantu memahami pola pembelian pelanggan terhadap menu-menu yang paling populer.

**Action:**
Berdasarkan pola pembelian tersebut, penjual dapat memanfaatkan menu yang sering dibeli dengan jumlah tinggi sebagai produk unggulan, misalnya dengan menjaga ketersediaan stok atau menjadikannya bagian dari paket promosi. Selain itu, jika terdapat menu yang biasanya dibeli dalam jumlah kecil, penjual bisa membuat strategi seperti paket bundling atau diskon pembelian lebih dari satu item agar pelanggan terdorong untuk membeli dalam jumlah lebih banyak.

## Scatter Plot

```python
plt.figure(figsize=(10, 6)) # Ukuran
plt.scatter(df['quantity'], df['transaction_amount'], alpha=0.7, color='teal') # Color (Warna)
plt.title('Hubungan Jumlah Item dan Nilai Transaksi', fontsize=16) # Judul
plt.xlabel('Jumlah Item yang Dibeli', fontsize=12) # X Keterangan
plt.ylabel('Jumlah Transaksi', fontsize=12) # Y Keterangan
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Insight:**
Scatter plot ini menunjukkan hubungan antara **harga item (item_price)** dan **jumlah pembelian (quantity)** dengan sangat rapih seperti diagonal sempurna. Dari grafik tersebut dapat dilihat bagaimana pola pembelian pelanggan terhadap harga produk, apakah produk dengan harga lebih murah cenderung dibeli dalam jumlah lebih banyak atau sebaliknya. Pola ini membantu memahami perilaku pelanggan dalam memilih produk berdasarkan harga.

**Action:**
Berdasarkan pola tersebut, penjual dapat menyesuaikan strategi penjualan dengan memperhatikan harga produk. Jika terlihat bahwa produk dengan harga lebih rendah lebih sering dibeli dalam jumlah banyak, maka produk tersebut bisa dijadikan fokus promosi atau paket penjualan. Selain itu, penjual juga dapat mempertimbangkan strategi diskon atau bundling pada produk tertentu agar pelanggan terdorong untuk membeli dalam jumlah lebih banyak.


## Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['item_price', 'quantity', 'transaction_amount']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('korelasi antara data numerik yang ada')
plt.show()
```

**Insight:**
Heatmap korelasi menunjukkan bahwa **item_price memiliki korelasi sangat kuat dengan transaction_amount**, yang berarti semakin tinggi harga suatu item maka nilai transaksi juga cenderung meningkat. Selain itu, quantity juga memiliki korelasi kuat dengan transaction_amount, sehingga semakin banyak item yang dibeli dalam satu transaksi maka total nilai transaksi akan semakin besar. Sementara itu, hubungan antara **item_price dan quantity relatif lemah**, yang menunjukkan bahwa harga produk tidak terlalu mempengaruhi jumlah item yang dibeli pelanggan.

**Action:**
Berdasarkan pola korelasi tersebut, penjual dapat meningkatkan pendapatan dengan mendorong pelanggan untuk membeli lebih banyak item dalam satu transaksi, misalnya melalui promo bundling atau paket hemat. Selain itu, karena harga item juga berkontribusi terhadap nilai transaksi, penjual dapat mempertimbangkan strategi penawaran pada produk dengan harga lebih tinggi atau membuat kombinasi produk agar pelanggan tetap tertarik membeli meskipun harga item berbeda.
